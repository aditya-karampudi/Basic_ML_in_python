#### With all features

```python
# Some data preprocessing
dataset = pd.get_dummies(dataset, columns=["Sex"])
```

```python
X = dataset.drop("age", axis=1)
y = dataset["age"]
```

```python
#splitting the data into train and test
from sklearn.cross_validation import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size=0.30)
```

```python
# import, instantiate, fit
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(xTrain, yTrain)
```

```python
# print the coefficients
print (linreg.intercept_)
print (linreg.coef_)
```

```python
#predict the test data
test_pred_lin = linreg.predict(xTest)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
#check the errors
rmse = sqrt(mean_squared_error(yTest, test_pred_lin))
print("rmse",rmse)
print("mean squared error",mean_squared_error(yTest, test_pred_lin))

print("mean absolute error",mean_absolute_error(yTest, test_pred_lin))
```

#### With few feature

```python
#change the attributes and try different combinations
features = ['Length', 'Height','Wholeweight']
X = dataset[features]
y = dataset['age']
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size=0.30)
```

```python
linreg = LinearRegression()
linreg.fit(xTrain, yTrain)
test_pred_lin = linreg.predict(xTest)
```

```python

rmse = sqrt(mean_squared_error(yTest, test_pred_lin))
print("rmse",rmse)
print("mean squared error",mean_squared_error(yTest, test_pred_lin))

print("mean absolute error",mean_absolute_error(yTest, test_pred_lin))
```
