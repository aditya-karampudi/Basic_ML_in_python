```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

# read data into dataset variable
data = pd.read_csv("Dataset_spine.csv")

data.head()
```

```python
data.shape
```

```python
data.columns = ['Pelvic Incidence','Pelvic Tilt','Lumbar Lordosis Angle','Sacral Slope','Pelvic Radius', 
                'Spondylolisthesis Degree', 'Pelvic Slope', 'Direct Tilt', 'Thoracic Slope', 
                'Cervical Tilt','Sacrum Angle', 'Scoliosis Slope','Outcome']

corr = data.corr()

# Set up the matplot figure
f, ax = plt.subplots(figsize=(12,9))

#Draw the heatmap using seaborn
sns.heatmap(corr, cmap='inferno', annot=True)
```

```python
data.describe()
```

```python
y =data['Outcome'].values
X = data.drop('Outcome',1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.cross_validation import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression(C=1)
```

```python
logreg.fit(xTrain, yTrain)
```

```python
y_log_pred_test = logreg.predict(xTest)
y_log_pred_train = logreg.predict(xTrain)
```

```python
metrics.confusion_matrix(yTest, y_log_pred_test)
```

```python
metrics.accuracy_score(yTest, y_log_pred_test)
```

```python
metrics.accuracy_score(yTrain, y_log_pred_train)
```

```python

#for grid search tuning C parameter and predicting the output
from sklearn.grid_search import GridSearchCV

log_cv = GridSearchCV(logreg, param_grid={'C':[1.2*i for i in range(1,6)], 'penalty':['l1','l2']}, cv=9)

log_cv.fit(xTrain, yTrain)
cv_predict = log_cv.predict(xTest)
cv_predict_train = log_cv.predict(xTrain)
```

```python
metrics.accuracy_score(cv_predict, yTest)
```

```python
metrics.accuracy_score(cv_predict_train, yTrain)
```

```python
log_cv.best_params_
#best parameters after tuning
```
